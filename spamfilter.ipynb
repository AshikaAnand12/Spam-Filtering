```python
import textract
import re
import numpy as np
import nltk
nltk.download('punkt')
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
```

```python
# Dictionary Address: Downloads/School/Advanced Data Mining/Homework/Spam Dictionary.pdf
dictIn = input("Enter Dictionary address:")
dictInput = str(textract.process(dictIn, method='pdfminer'))
dictInput = dictInput.lower()
arrDict = dictInput.split('\\n')
spamDict = []
for element in arrDict:
    spamDict.append(element.strip(" "))
spamDict = spamDict[2:]
spamDict = spamDict[:-3]
print(spamDict)
```

```python
# File Address: Downloads/School/Advanced Data Mining/Homework/Document1.pdf
i=0
spamDoc="default"
arr = []
while(True):
    spamDoc = input("Enter document address:")
    arrIndx = 0
    if (spamDoc == ""):
        break
    unprocessed_text = str(textract.process(spamDoc, method='pdfminer')).replace("\\n"," ")
    processed = unprocessed_text
    char1 = '>'
    if char1 in unprocessed_text:
        arrIndx = unprocessed_text.rindex('>')+1
        processed = unprocessed_text[arrIndx:]
    char2 = 'CMPE'
    if char2 in processed:
        arrIndx = processed.rindex('CMPE')
        processed = processed[:arrIndx]
    processed = processed.strip().lower()
    doc = []
    n=0
    for item in spamDict:
        n = processed.count(item)
        doc.append(n)
    arr.append(doc)
print("Term frequency matrix: ",arr)
```

```python
doc_freq = []
for i in arr:
    n = 0
    for ele in i:
        if(ele!=0):
            n=n+1
    doc_freq.append(n)
tfidf_val = []
for i in arr:
    val = []
    for item in i:
        if(item!=0):
            tf = 1 + np.log(1+np.log(item))
            idf = np.log((1+len(i))/doc_freq[arr.index(i)])
            compute = tf*idf
        else:
            tf = 0
            tf_idf = 0
            compute = 0
        val.append(compute)
    tfidf_val.append(val)
print("TF-IDF values = ",tfidf_val)
```

```python
rank=[]
for oned in tfidf_val:
    ranking = 0
    for val in oned:
        ranking = ranking + val
    rank.append(ranking)
print(rank)
```
